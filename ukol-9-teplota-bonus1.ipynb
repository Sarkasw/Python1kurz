```python
import pandas
import pytemperature

teploty = pandas.read_csv("temperature.csv")
teploty
```

```python
teploty["AvgCelsia"] = pytemperature.f2c(teploty["AvgTemperature"])
teploty
```

```python
nad_30 = teploty[teploty["AvgCelsia"] > 30]
nad_30
```

```python
nad_15_Evropa = teploty[(teploty["AvgCelsia"] > 15) & (teploty["Region"] == "Europe")]
nad_15_Evropa
```

```python
extremni_mereni = teploty[(teploty["AvgCelsia"] > 30) | (teploty["AvgCelsia"] < -10)]
extremni_mereni.sort_values("AvgCelsia").tail(50)
```
