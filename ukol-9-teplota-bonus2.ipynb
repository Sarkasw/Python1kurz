```python
import pandas
import pytemperature

teploty = pandas.read_csv("temperature.csv")
teploty
```

```python
teploty["AvgCelsia"] = pytemperature.f2c(teploty["AvgTemperature"])
teploty
```

```python
trinact = teploty[teploty["Day"] == 13]
trinact
```

```python
trinact_usa = teploty[(teploty["Day"] == 13) & (teploty["Country"] == "US")]
trinact_usa
```

```python
was_a_pil = trinact_usa[(trinact_usa["City"] == "Washington") | (trinact_usa["City"] == "Philadelphia")]
was_a_pil
```
