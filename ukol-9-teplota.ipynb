```python
import pandas
teploty = pandas.read_csv("temperature.csv")
teploty
```

```python
mereni_praha = teploty[teploty["City"] == "Prague"]
mereni_praha
```

```python
nad_80 = teploty[teploty["AvgTemperature"] > 80]
nad_80
```

```python
nad_60_evropa = teploty[(teploty["AvgTemperature"] > 60) & (teploty["Region"] == "Europe")]
nad_60_evropa
```

```python
teplotni_extremy = teploty[(teploty["AvgTemperature"] > 80) | (teploty["AvgTemperature"] < -20)]
teplotni_extremy.sort_values("AvgTemperature").head(60)
```
