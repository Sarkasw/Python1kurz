```python
teploty = pd.read_csv("temperature.csv")
mereni_helsinki = teploty[teploty["City"] == "Helsinki"]
mereni_helsinki = mereni_helsinki[["Day","AvgTemperature"]]
mereni_miami = teploty[teploty["City"] == "Miami Beach"]
mereni_miami = mereni_miami[["Day","AvgTemperature"]]
mereni_tokyo = teploty[teploty["City"] == "Tokyo"]
mereni_tokyo = mereni_tokyo[["Day","AvgTemperature"]]
mereni = pd.merge(mereni_helsinki, mereni_miami, on="Day")
mereni = pd.merge(mereni, mereni_tokyo, on="Day")
mereni = mereni.set_index('Day')
mereni = mereni.rename(columns={"Day": "Den", "AvgTemperature_x": "Helsinki", "AvgTemperature_y": 'Miami Beach', "AvgTemperature": 'Tokyo'})
mereni
```

```python
mereni.plot(kind="box")
```
